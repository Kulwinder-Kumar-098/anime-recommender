```python
import os
anime=pd.read_csv(os.path.join(cooperunion_anime_recommendations_database_path, 'anime.csv'))
rating=pd.read_csv(os.path.join(cooperunion_anime_recommendations_database_path, 'rating.csv'))
```

```python
anime['rating']=anime['rating'].fillna(anime['rating'].mean())
```

```python
anime=anime.dropna()
```

```python
anime['tags']=anime['genre']+","+anime['type']
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(obj):
    y=[]
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('adventure')
```

```python
new_anime=anime[['anime_id','name',	'tags']]
```

```python
new_anime['tags']=new_anime['tags'].apply(stem)
```

```python
new_anime.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
```

```python
vectors=cv.fit_transform(anime['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
reco_anime=new_anime[new_anime['name']=='Fullmetal Alchemist: Brotherhood'].index[0]
```

```python
def recommend_anime_smart(user_input, top_n=5):
    print(f"Searching for recommendations for: '{user_input}'")

    # Try to find an exact match for an anime name (case-insensitive)
    matched_anime = new_anime[new_anime['name'].str.lower() == user_input.lower()]

    if not matched_anime.empty:
        anime_index = matched_anime.index[0]
        distances = similarity[anime_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:top_n + 1] # +1 to exclude itself
        print(f"Recommendations based on anime '{matched_anime.iloc[0]['name']}':")
    else:
        print(f"Anime '{user_input}' not found. Providing recommendations based on input description.")
        stemmed_input = stem(user_input.lower())
        input_vector = cv.transform([stemmed_input]).toarray()
        distances = cosine_similarity(input_vector, vectors)[0]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[:top_n]
        print(f"Recommendations based on description '{user_input}':")

    for i, score in movies_list:
        print(f"- {new_anime.iloc[i]['name']} (Similarity: {score:.2f})")
```

Now, let's try the new recommendation function with a custom input.

```python
recommend_anime_smart('naruto')
print("\n")
recommend_anime_smart('Fullmetal Alchemist: Brotherhood')
print("\n")
recommend_anime_smart('Sci-fi thriller movie about space')
print("\n")
recommend_anime_smart('')
```
